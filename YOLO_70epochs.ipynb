```python
import zipfile
import os
import shutil
import random
import matplotlib.pyplot          as plt
from ultralytics          import YOLO
from google.colab         import drive
```

##BUILD ROBOCUP_2019 Dataset

###Dataset details
RoboCup Dataset contains 196195 images.

8 main parent categories that contain 180 children categories.

80% for train

20% for valid

###Build dataset

###Connect to my oneDrive

```python
#drive.mount('/content/drive')
ruta_archivo_zip = '/content/drive/MyDrive/Colab Notebooks/data_subclass.zip'
ruta_destino = './'
with zipfile.ZipFile(ruta_archivo_zip, 'r') as archivo_zip:
    # Extraer el archivo deseado en la ruta de destino
    archivo_zip.extractall(ruta_destino)
```

###transforming data to suit the yolo model

```python
def transform_dataset(root_dir, output_dir, train_ratio=0.8):
    # Create the output directory
    os.makedirs(output_dir, exist_ok=True)
    train_dir = os.path.join(output_dir, 'train')
    val_dir = os.path.join(output_dir, 'val')
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # Get the list of classes in the root directory
    classes = [d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))]

    for class_name in classes:
        class_dir = os.path.join(root_dir, class_name)
        if not os.path.isdir(class_dir):
            continue

        # Get the list of images in the current class
        images = [f for f in os.listdir(class_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # Shuffle the images randomly
        random.shuffle(images)

        # Calculate the number of training images based on the specified ratio
        train_size = int(len(images) * train_ratio)

        # Create the class directories in the training and validation sets
        train_class_dir = os.path.join(train_dir, class_name)
        val_class_dir = os.path.join(val_dir, class_name)
        os.makedirs(train_class_dir, exist_ok=True)
        os.makedirs(val_class_dir, exist_ok=True)

        # Copy the training images
        for img in images[:train_size]:
            src_path = os.path.join(class_dir, img)
            dst_path = os.path.join(train_class_dir, f"{os.path.splitext(img)[0]}_{class_name}.png")
            shutil.copy(src_path, dst_path)

        # Copy the validation images
        for img in images[train_size:]:
            src_path = os.path.join(class_dir, img)
            dst_path = os.path.join(val_class_dir, f"{os.path.splitext(img)[0]}_{class_name}.png")
            shutil.copy(src_path, dst_path)
```

```python
root_dir   = '/content/training_data'   # Path to the original dataset
output_dir  = '/content/dataset'      # Path for the resized dataset

transform_dataset(root_dir, output_dir, train_ratio=0.8)
```

##YOLO

###Train

```python
# Load a model
model = YOLO('yolov8x-cls.yaml').load('yolov8x-cls.pt')  # build from YAML and transfer weights

# Train the model
model.train(data='/content/dataset', epochs=70, patience=70, imgsz=224,
      verbose=True, batch=64,
      project='/content/drive/MyDrive/Colab Notebooks/save_models', name='yolo_70epoch/classify')
```

###Val

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.top1   # top1 accuracy
metrics.top5   # top5 accuracy
```

##YOLO Superclass

###Build dataset

###Connect to my oneDrive

```python
#drive.mount('/content/drive')
ruta_archivo_zip = '/content/drive/MyDrive/Colab Notebooks/data_superclass.zip'
ruta_destino = './'
with zipfile.ZipFile(ruta_archivo_zip, 'r') as archivo_zip:
    # Extraer el archivo deseado en la ruta de destino
    archivo_zip.extractall(ruta_destino)
```

###transforming data to suit the yolo model

```python
def transform_dataset(root_dir, output_dir, train_ratio=0.8):
    # Crear el directorio de salida
    os.makedirs(output_dir, exist_ok=True)
    train_dir = os.path.join(output_dir, 'train')
    val_dir = os.path.join(output_dir, 'val')
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # Obtener la lista de clases en el directorio ra√≠z
    classes = [d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))]

    for class_name in classes:
        class_dir = os.path.join(root_dir, class_name)
        subclasses = [d for d in os.listdir(class_dir) if os.path.isdir(os.path.join(class_dir, d))]

        for subclass_name in subclasses:
            subclass_dir = os.path.join(class_dir, subclass_name)

            # Obtener la lista de im√°genes en la subclase actual
            images = [f for f in os.listdir(subclass_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

            # Mezclar las im√°genes al azar
            random.shuffle(images)

            # Calcular el tama√±o del conjunto de entrenamiento basado en el ratio especificado
            train_size = int(len(images) * train_ratio)

            # Crear los directorios de clase en los conjuntos de entrenamiento y validaci√≥n
            train_class_dir = os.path.join(train_dir, class_name)
            val_class_dir = os.path.join(val_dir, class_name)
            os.makedirs(train_class_dir, exist_ok=True)
            os.makedirs(val_class_dir, exist_ok=True)

            # Copiar las im√°genes de entrenamiento
            for img in images[:train_size]:
                src_path = os.path.join(subclass_dir, img)
                dst_path = os.path.join(train_class_dir, f"{os.path.splitext(img)[0]}_{class_name}.png")
                shutil.copy(src_path, dst_path)

            # Copiar las im√°genes de validaci√≥n
            for img in images[train_size:]:
                src_path = os.path.join(subclass_dir, img)
                dst_path = os.path.join(val_class_dir, f"{os.path.splitext(img)[0]}_{class_name}.png")
                shutil.copy(src_path, dst_path)
```

```python
root_dir   = '/content/dataset_superclass'   # Path to the original dataset
output_dir  = '/content/dataset_super'      # Path for the resized dataset

transform_dataset(root_dir, output_dir, train_ratio=0.8)
```

###Train

```python
# Load a model
model = YOLO('yolov8x-cls.yaml').load('yolov8x-cls.pt')  # build from YAML and transfer weights

# Train the model
model.train(data='/content/dataset_super', epochs=70, patience=70, imgsz=224,
      verbose=True, batch=64,
      project='/content/drive/MyDrive/Colab Notebooks/save_models', name='yolo_70epoch_superClass/classify')
```

###Val

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.top1   # top1 accuracy
metrics.top5   # top5 accuracy
```
